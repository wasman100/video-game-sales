```python
import pandas as pd
import matplotlib.pyplot as plt
```

The code below is used to resize later graphs and to call the csv file. It also replace the string tbd also replacing all the NA with the value 0 and then convert the User_Score column to be that of an int so it is easier to compare the columns

```python
plt.rcParams['figure.figsize'] = (20,5)
sales=pd.read_csv('data/Video_Games_Sales_as_at_22_Dec_2016.csv',delimiter = ',')
sales.User_Score =sales.User_Score.replace("tbd", 0)
sales.fillna(0, inplace = True)
sales.User_Score=pd.to_numeric(sales.User_Score)
```

The below code is taking only the columns we want to compare to make it easier on us down the line for sorting and graphing

```python
bar_compare= sales[['Critic_Score','NA_Sales','User_Score']].copy()
```

the below graphs critic score vs na sales and then graphs user score vs na sales 

```python
bar_compare.sort_index().plot.scatter(x="Critic_Score", y="NA_Sales", color='red');
bar_compare.sort_index().plot.scatter(x="User_Score", y="NA_Sales");
```
